```python
import os
import pandas as pd


print("Files in /content directory:")
print(os.listdir('/content'))

import pandas as pd

df = pd.read_csv('/content/netflix_titles.csv')

# 1. How many rows and columns are cointained in the dataset?
print(f"1. Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# 2. What datatypes are contained in the dataset?
print("\n2. Data types:")
print(df.dtypes)

# 3. Is the dataset complete? i.e no missing values?
print("\n3. Missing values:")
missing_values = df.isnull().sum()
print(missing_values)
print(f"Dataset is {'complete' if missing_values.sum() == 0 else 'incomplete'}")

# 4. List the number of Movies and TV shows among the last 20 rows of the dataset
print("\n4. Last 20 rows - Type distribution:")
last_20 = df.tail(20)
print(last_20['type'].value_counts())


# 1. Give the number of all Movies originating from South Africa.
sa_movies = df[(df['type'] == 'Movie') & (df['country'] == 'South Africa')]
print(f"1. Movies from South Africa: {len(sa_movies)}")

# 2. What is the description of the movies in the 12th row?
print(f"\n2. Description of movie in 12th row:")
print(df.iloc[11]['description'])

# 3. From these South African movies, list only those released in 2021.
sa_2021 = sa_movies[sa_movies['release_year'] == 2021]
print(f"\n3. South African movies from 2021: {len(sa_2021)}")
if len(sa_2021) > 0:
    print(sa_2021[['title', 'release_year']].to_string(index=False))

# 4. List the entire cast of all movies released in 2021.
movies_2021 = df[(df['type'] == 'Movie') & (df['release_year'] == 2021)]
print(f"\n4. Cast of movies from 2021 ({len(movies_2021)} movies):")
for idx, row in movies_2021.iterrows():
    print(f"\nMovie: {row['title']}")
    print(f"Cast: {row['cast'] if pd.notna(row['cast']) else 'Not available'}")
```
